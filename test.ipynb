```python
selected_columns = df_historical.columns.difference(["datetimes"])
df_historical.loc[:,selected_columns].describe().T
```

```python
df_historical['years'] = df_historical['datetimes'].dt.year
df_historical['months'] = df_historical['datetimes'].dt.strftime('%B')
df_historical['days'] = df_historical['datetimes'].dt.day

# Sütun sırasını değiştirmek için sütun isimlerini kullanmadan, sadece sırayı belirtmek:
df_historical = df_historical[['datetimes', 'years','months','days'] + [col for col in df_historical.columns if col not in ['datetimes', 'years','months','days']]]
```
